```python
import pandas as pd
import numpy as np
import ast 
```

```python
#Dirección del archivo
direccion = r"C:\Users\Elias\Documents\Tercer proyecto individual\Datasets\credits.csv"

# Lectura del archivo CSV
datos = pd.read_csv(direccion)

# Convertimos en un DataFrame
creditos = pd.DataFrame(datos)

# Mostramos las primeras 5 filas del DataFrame
creditos.head()
```

```python
# Creamos un dataframe a partir de la primera columna ["cast"] y con la columna ["id"]
cast_df = creditos[["cast", "id"]]
# Creamos un dataframe a partir de la primera columna ["crew"] y con la columna ["id"]
crew_df = creditos[["crew", "id"]]
# Visualizamos los datos del 1er dataframe
cast_df.head()
```

```python
# Creamos una función para desanidar las columnas
def desanidar(df, columns):
    for column in columns:
        df[column] = df[column].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
        # Expandir la columna en filas
        df = df.explode(column)
        # Normalizar la columna expandida
        col_df = pd.json_normalize(df[column])
        # Renombrar las columnas normalizadas para evitar superposiciones
        col_df = col_df.add_prefix(f'{column}')
        # Concatenar las columnas normalizadas con el DataFrame original
        df = df.drop(columns=[column]).reset_index(drop=True).join(col_df)
    return df
```

```python
# Aplicamos la función desanidar a lso 2 dataframes
cast_df_desanidado = desanidar(cast_df, ["cast"])
crew_df_desanidado = desanidar(crew_df, ["crew"])
```

```python
# Visualizamos el dataframe "cast"
cast_df_desanidado.head()
```

```python
# Visualizamos el dataframe "crew"
crew_df_desanidado.head()
```

```python
# Escogemos las columnas que ocuparemos para nuestro modelo, del dataframe "cast"
actores = cast_df_desanidado[["id","castname"]].copy().rename(columns={"castname":"Actor"})

# Reseteamso el indice
actores.reset_index(drop=True, inplace=True)

# Visualizamos los resultados del dataframe "actores"
actores.head()
```

```python
# Escogemos las columnas que ocuparemos para nuestro modelo, del dataframe "crew"
crewjobs = crew_df_desanidado['crewjob'].unique()

# Visualizamos
crewjobs
```

```python
# Creamos una mascara que nos permita obtener los valores en los que crewjob sea == "Director"
directores = crew_df_desanidado[crew_df_desanidado['crewjob'] == 'Director']

# Aplicamos la mascara que nos permita obtener los valores
directores = directores[["id", "crewname"]].copy().rename(columns={"crewname": "Director"})

# Reseteamos el indice
directores = directores.reset_index(drop=True)

# Visualizamos los resultados del dataframe "directores"
directores.head(3)
```

```python
# Dado que un director puede dirigir multiples peliculas o una pelicual puede tener multiples directores verificamos si tenemos peliculas asi
crew_df_desanidado[(crew_df_desanidado['id'] == 5) & (crew_df_desanidado['crewjob'] == 'Director')]
```

```python
# Unimos los 2 dataframes por id , usando merge 
peliculas = pd.merge(actores, directores, on="id", how="left")

# Visualizamos los resultados del dataframe "peliculas"
peliculas.head(3)
```

```python
# Convertimos la columna ["Actor"] y la columna ["Director"] en cadenas de textos
peliculas['Actor'] = peliculas['Actor'].astype(str)
peliculas['Director'] = peliculas['Director'].astype(str)
```

```python
# Agrupamos las cadenas
creditos = peliculas.groupby('id').agg({
     #Aquí se verificará si hay más de un director después de convertirse en set (eliminándose duplicados). Si hay más de 1, se concatenan separados por comas. Si sólo hay uno, se agrega el primer valor.
    'Director': lambda x: ', '.join(set(x)) if len(set(x)) > 1 else x.iloc[0], 
    'Actor': lambda x: ', '.join(x) #En este caso, siempre se concatenan los elementos.
}).reset_index() #Se concatenan por 'id' de película a los directores y actores
creditos.head()
```

```python
# Por ultimo verificamos que no hallan duplicados
creditos.drop_duplicates(inplace=True)
```

```python
# Guardamos el dataset como archivo parquet
creditos.to_parquet(r'C:\Users\Elias\Documents\Tercer proyecto individual\Datasets_limpios\creditos_desanidados.parquet')
```
