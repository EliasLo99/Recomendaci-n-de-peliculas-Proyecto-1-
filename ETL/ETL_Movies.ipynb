```python
import pandas as pd
import numpy as np
```

```python
# Ubicación del archivo
direccion = r"C:\Users\Elias\Documents\Tercer proyecto individual\Datasets\movies_dataset.csv"

# Lectura del dataset
data = pd.read_csv(direccion)

# Convertimos en un DataFrame
movies = pd.DataFrame(data)

# Mostramos las primeras 5 lineas del dataframe
movies.head()
```

```python
#Eliminamos las filas que sean completamente nulos
movies = movies.dropna(how='all')

# Eliminamos las columnas del dataframe que tengan un gran porcentaje de valores nulos o que no ocuparemos
movies = movies.drop(columns=["adult","video","poster_path","homepage","imdb_id","original_title"])

# Revisamos la cantidad de filas y columnas después de la eliminación
movies.head()
```

```python
# Buscamos los valores en el DataFrame que esten vacios 
null_values = movies.isnull().sum()

# Mostramos los valores nulos
null_values
```

```python
#Remplazamos los valores nulos de la columna ["belongs_to_collection"]
movies['belongs_to_collection'] = movies['belongs_to_collection'].fillna('NaN')
movies['belongs_to_collection'].head()
```

```python
# Observamos el 1er valor de la fila
movies['belongs_to_collection'][0]
```

```python
# Función para verificar y desanidar
def unpack(value):
    if isinstance(value, str) and len(value) > 1:  # Verifica que sea string y longitud > 1
        return value[1:-1]  # Devuelve la cadena sin el primer y último carácter
    else:
        return 'No founder'

# Aplica la función a la columna 'belongs_to_collection'
movies['belongs_to_collection'] = movies['belongs_to_collection'].apply(unpack)
```

```python
# Convertimos los valores de la columna belongs_to_collection en lista
movies["belongs_to_collection"] = [item.replace("'", "") for item in movies["belongs_to_collection"]]

# Separamos los valores de la columna en sub strings por medio de la  ","
movies['belongs_to_collection'] = movies['belongs_to_collection'].apply(lambda x: x.split(','))

# Mostramos los valores de la columna
movies['belongs_to_collection'].head()
```

```python
# Creamos columnas por cada pocision de la lista que esta en la columna ["belongs_to_collection"]
for i in range(len(movies['belongs_to_collection'][0])):
    movies[f'collection_{i+1}'] = [item[i] if i < len(item) else 'No Collection' for item in movies['belongs_to_collection']]
```

```python
# Eliminamos las filas que no ocuparemos o hemos dejado de usar 
movies = movies.drop(columns=["belongs_to_collection","collection_3","collection_4"])

# Renombramos las columnas collection_1 y collection_2
movies.rename(columns={"collection_1":"id_collection", "collection_2":"name_collection"}, inplace=True)
```

```python
# Eliminamos la columna ["id_collection"]
movies = movies.drop(columns=["id_collection"])
```

```python
# Función para procesar id_collection
def proceso(value):
    # Verifica que el valor no sea nulo y contenga el carácter ':'
    if isinstance(value, str) and ':' in value:
        return value.split(':')[1].strip()  # Toma el segundo elemento y elimina espacios
    else:
        return value  # Retorna el valor original si no cumple la condición


# Aplicamos la funcion en la columna "name_collection"
movies['name_collection'] = movies['name_collection'].apply(proceso)
```

```python
# Visualizamos el dataframe con las nuevas columnas agregadas
movies.head(3)
```

```python
# Analizamos la columna ["production_companies"]
movies["production_companies"][0]

# Aplica la función a la columna 'production_companies'
movies['production_companies'] = movies['production_companies'].apply(unpack)

# Convertimos los valores de la columna production_companies en lista
movies["production_companies"] = [item.replace("'", "") for item in movies["production_companies"]]

# Separamos los valores de la columna en sub strings por medio de la  ","
movies['production_companies'] = movies['production_companies'].apply(lambda x: x.split(','))
```

```python
# convertimos en columna solamente el primer subpocision de la fila de la columna production_companies
for i in range(len(movies['production_companies'][1])):
    movies[f'production_company_{i+1}'] = [item[i] if i < len(item) else 'No Production Company' for item in movies['production_companies']]

# Observamos el dataframe con las nuevas columnas
movies.head(2)
```

```python
# Aplicamos la función proceso en la columna production_company_1 para sacar el nombre de la compañia
movies['production_company_1'] = movies['production_company_1'].apply(lambda x: x.split('name: ')[1] if isinstance(x, str) and 'name:'in x else 'No Production Company')

# Eliminamos los espacios vacios de las filas
movies['production_company_1'] = movies['production_company_1'].str.strip()

# Observamos La columna con el 1er valor
movies['production_company_1'].head()
```

```python
# Eliminamos las filas que no ocuparemos o hemos dejado de usar
movies = movies.drop(columns=["production_companies","production_company_2","production_company_3","production_company_4","production_company_5","production_company_6"])

# Renombramos las columnas production_company_1
movies.rename(columns={"production_company_1":"production_company"}, inplace=True)
```

```python
# Eliminamos los valores nulos en la columna release_date 
movies = movies.dropna(subset=['release_date'])

# Eliminar filas con fechas que no cumplen el formato datetime
movies = movies[movies['release_date'].str.match(r'\d{4}-\d{2}-\d{2}', na=False)]

# Creamos una columna llamada años, a partir de la columna relase_date
movies['release_date'] = pd.to_datetime(movies['release_date'])

#creamos una columna con el años, a partir de la columna relase_date 
movies['years'] = movies['release_date'].dt.year
```

```python
# Remplazamos los valores nulos pór 0 en la columna revenue 
movies['revenue'] = movies['revenue'].fillna(0)

# Remplazamos los valores nulos por 0 en la columna budget
movies['budget'] = movies['budget'].fillna(0)
```

```python
# Convertimos la columna ["revenue"] en numerica
movies['revenue'] = pd.to_numeric(movies['revenue'], errors='coerce')

# Convertimos la columna ["budget"] en numerica
movies['budget'] = pd.to_numeric(movies['budget'], errors='coerce')

# Calcular el retorno de inversión, asignando 0 cuando no haya datos disponibles
movies['return_on_investment'] = np.where(
    (movies['revenue'].isna()) | (movies['budget'].isna()) | (movies['budget'] == 0),
    0,
    movies['revenue'] / movies['budget'] )
```

```python
# Visualizamos como va evoluciónando el dataframe
movies.head(3)
```

```python
# Desenpaquetamos la columna ["genres"]
movies['genres'] = movies['genres'].apply(unpack)

# Convertimos los valores de la columna genres en lista
movies["genres"] = [item.replace("'", "") for item in movies["genres"]]

# Separamos los valores de la columna en sub strings por medio de la  ","
movies['genres'] = movies['genres'].apply(lambda x: x.split(','))
```

```python
# Creamos una columna por cada pocision de la fila , en la columna ["genres"]
for i in range(len(movies['genres'][1])):
    movies[f'genre_{i+1}'] = [item[i] if i < len(item) else 'No Genre' for item in movies['genres']]

# Observamos las nueevas columnas
movies.head(2)
```

```python
# Eliminamos las columnas que no ocuparemos o ya no son indispensables 
movies = movies.drop(columns=["genres","genre_1","genre_3","genre_4","genre_5","genre_6"])

# Renombramos las columnas
movies.rename(columns={"genre_2":"genre"}, inplace=True)

# Visualizamos la columna con los géneros
movies['genre'].head()
```

```python
# Aplicamos la función proceso a la columna ["genre"]
movies['genre'] = movies['genre'].apply(proceso)

# Eliminamos el ultimo valor de la fila en la columna
movies['genre'] = movies['genre'].apply(lambda x: x[:-1] if isinstance(x, str) else 'No Genre')

# Eliminamos los espacios vacios de las filas
movies['genre'] = movies['genre'].str.strip()
```

```python
# Visualizamos la columna
movies['genre'].head()
```

```python
# Convertimos la columna ["popularity"] en categoria numerica# Convertimos la columna 'popularity' a float, reemplazando valores no numéricos con NaN y luego con 0
movies['popularity'] = pd.to_numeric(movies['popularity'], errors='coerce').fillna(0)

# Guardamos el DataFrame en formato Parquet
movies.to_parquet('movies_clean.parquet')

```
